```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#  Прочитали данные
data = pd.read_csv('Automobile_data.csv')
data
```

```python
#  Группировка по маркам авто, расчет среднего для оставшихся столбцов внутри групп, вывод только company (index) и price
temp_table = data.groupby('company').mean()['price']
#  Получение списка значений из индекса
groups = temp_table.index.tolist()
#  Получение списка значений для оставшегося столбца price
counts = temp_table.tolist()
plt.bar(groups, counts)
```

```python
#  Столбец с индексами компаниями и долями компаний, округлынными до 2 знаков
temp_table_2 = round(data.groupby('company').size() / len(data) * 100, 2)

#  Получение списка значений столбца
vals = temp_table_2.tolist()
#  Получение списка значений из индекса (компании)
labels = temp_table_2.index.tolist()

fig, ax = plt.subplots()
ax.pie(vals, labels=labels, autopct='%1.1f%%')
ax.axis("equal")
```

```python
#  Столбец с индексом (тип кузова) и долями типов кузова, округлынными до 2 знаков
temp_table_2 = round(data.groupby('body-style').size() / len(data) * 100, 2)

#  Получение списка значений столбца
vals = temp_table_2.tolist()
#  Получение списка значений из индекса (тип кузова)
labels = temp_table_2.index.tolist()

fig, ax = plt.subplots()
ax.pie(vals, labels=labels, autopct='%1.1f%%')
ax.axis("equal")
```

```python
#  Столбец с индексом (тип двигателя) и долями типов двигателя, округлынными до 2 знаков
temp_table_2 = round(data.groupby('engine-type').size() / len(data) * 100, 2)

#  Получение списка значений столбца
vals = temp_table_2.tolist()
#  Получение списка значений из индекса (тип двигателя)
labels = temp_table_2.index.tolist()

fig, ax = plt.subplots()
ax.pie(vals, labels=labels, autopct='%1.1f%%')
ax.axis("equal")
```
